```python
import numpy as np

def damped_harmonic_oscillator_euler(x0, v0, w, d, dt, t_max):
  """
  Solves the damped harmonic oscillator equation using Euler's method.

  Args:
      x0: Initial position.
      v0: Initial velocity.
      w: Natural frequency.
      d: Damping constant.
      dt: Timestep.
      t_max: Maximum time.

  Returns:
      t: Array of time points.
      x: Array of positions at each time point.
      v: Array of velocities at each time point.
  """
  t = np.arange(0, t_max + dt, dt)
  x = np.zeros_like(t)
  v = np.zeros_like(t)

  x[0] = x0
  v[0] = v0

  for i in range(1, len(t)):
    a = -2 * d * v[i-1] - w**2 * x[i-1]
    v[i] = v[i-1] + dt * a
    x[i] = x[i-1] + dt * v[i]

  return t, x, v

x0 = 1.0
v0 = 0.0
w = 2.0 * np.pi
d = 0.5
dt = 0.01
t_max = 10.0

t, x, v = damped_harmonic_oscillator_euler(x0, v0, w, d, dt, t_max)
```

```python
import matplotlib.pyplot as plt

plt.plot(t, x, label='Position')
plt.plot(t, v, label='Velocity')
plt.xlabel('Time (t)')
plt.ylabel('Position (x) or Velocity (v)')
plt.legend()
plt.show()
```

```python
w_values = [1.0, 2.0, 3.0]
d_values = [1, 0.5, 0.8] 

for w in w_values:
    for d in d_values:
        t, x, v = damped_harmonic_oscillator_euler(x0, v0, w, d, dt, t_max)
        label = f'w={w}, d={d}'
        plt.plot(t, x, label=label)

plt.xlabel('Time (t)')
plt.ylabel('Position (x)')
plt.legend()
plt.show()
```
