```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, 3], [3, -1], [2, 2]])
b = np.array([-2, 4, 1])


def calculate_sse(x):
  """Calculates the sum of squared residuals for a given solution vector x."""
  predictions = A.dot(x)
  errors = predictions - b
  return np.sum(errors**2)

def visualize_lines(points, title):
  """Visualizes the lines and points in a 2D plot."""
  x_range = np.linspace(-5, 5, 100)
  for point in points:
    y1 = (-point[0] * x_range + point[1]) / point[1] 
    plt.plot(x_range, y1)
  plt.title(title)
  plt.xlabel("x1")
  plt.ylabel("x2")
  plt.legend()
  plt.grid(True)
  plt.show()

sample_solutions = np.array([[-1.2, 0.8], [-0.8, 1.2]])

visualize_lines(A, "Original Equations")

for solution in sample_solutions:
  plt.scatter(solution[0], solution[1], color='red', marker='o', label='Sample Solution')
visualize_lines(A, "Equations with Sample Solutions")
pinv_A = np.linalg.pinv(A)
x_pinv = pinv_A @ b
print("Solution using pseudoinverse (x):", x_pinv)

plt.scatter(x_pinv[0], x_pinv[1], color='green', marker='o', label='Pseudoinverse Solution')
visualize_lines(A, "Equations with Pseudoinverse Solution")
```
