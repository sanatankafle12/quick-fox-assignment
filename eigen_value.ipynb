```python
import numpy as np
```

```python
A = np.array([[40,-20,0],
              [-20,40,-20],
              [0,-20,40]])
```

```python
def power_method_iteration(A, tolerable_error = 1e-6, maximum_iteration = 100):
    """
    Calculate Eigen Value, Eigen Vector and Error

    Args:
        A : Square Matrix
        tolerable_error : Level chekcing for convergence
        maximum_iteration : Maximum number of iterations

    Output:
        eigenvalue : Largest EigenValue -> Float
        eigenvector : Corresponding Vector -> Numpy Array
        errors : List of differences between current and previous Eigen value -> List
    """

    #Initial Guess
    x = [1,1,1]
    errors = []

    for i in range(maximum_iteration):
        x_prev = x.copy()
        x = np.dot(A,x)
        norm = np.linalg.norm(x)
        x = x / norm
        error = np.linalg.norm(x-x_prev)
        errors.append(error)
        if error < tolerable_error:
            break
    eigenvalue = np.dot(np.dot(A.T, x),x)
    return eigenvalue, x, errors,i

```

```python
eigenvalue, eigenvector, errors, i = power_method_iteration(A)
```

```python
print(f"The largest Eigen Values is {eigenvalue}.\nThe corresponding Eigen Vector is {eigenvector}.\nThe Errors are {errors}.")
```

```python
import matplotlib.pyplot as plt

plt.plot(range(len(errors)),errors)
plt.show()
```

### As we can see, the value λ Converges.

This is because, we can see that error in the eigenvalues decreases as the iterations progress. In this case, the error decreased from ~1.082 to ~6.22e-07 after a certain number of iterations. The decreasing trend suggests that the estimated eigen value is getting closer to the original one. 

### Significance Associated with Largest eigenvalue

1. Eigen Values allows us to understand the behaviour of a system and make predictions 
2. In case of PCA(Principle Component Analysis), Eigen values represent total amount of variance. It gives us idea about the most important factor in the dataset
